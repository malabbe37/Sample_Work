```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

NHANES = pd.read_csv(r'C:\Users\malab\OneDrive\Documents\Notre Dame\Introduction to Data Science\Data Sets\NHANES.csv')
NHANES.info()
```

```python
NHANES2 = NHANES[NHANES.Age > 18]
NHANES2 = NHANES2[["HardDrugs","RegularMarij","Age","AlcoholYear","BMI","Gender","HHIncomeMid","Weight","Height","SexNumPartnLife","TotChol"]]
NHANES2 = NHANES2.dropna()
NHANES2.info()
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(NHANES2,train_size=.8,random_state=1842)
train.info()
```

```python
from sklearn.compose import ColumnTransformer, make_column_transformer, make_column_selector
from sklearn.preprocessing import OneHotEncoder
my_ct=make_column_transformer((OneHotEncoder(drop="if_binary"),["RegularMarij","Gender"]),remainder="passthrough")
train2=pd.DataFrame(my_ct.fit_transform(train))
train2.columns = ['RegularMarij','Gender','HardDrugs','Age','AlcoholYear','BMI','HHIncomeMid','Weight','Height','SexNumPartnLife','TotChol']
train2.head()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
NHANES_X=NHANES2[["RegularMarij","Age","AlcoholYear","BMI","Gender","HHIncomeMid","Weight","Height","SexNumPartnLife","TotChol"]]
NHANES_y=NHANES2["HardDrugs"]

X_train,X_test,y_train,y_test=train_test_split(NHANES_X,NHANES_y,test_size=0.2,random_state=1842)
```

```python
my_ct=make_column_transformer((OneHotEncoder(drop="if_binary"),["RegularMarij","Gender"]),remainder="passthrough")
X_train2=pd.DataFrame(my_ct.fit_transform(X_train))
X_train2.columns = ['RegularMarij','Gender','Age','AlcoholYear','BMI','HHIncomeMid','Weight','Height','SexNumPartnLife','TotChol']
X_train2.head()
```

```python
my_rsearch=RandomizedSearchCV(
    estimator=RandomForestClassifier(),
    param_distributions={"max_features":[2,3,4,5,6,7,8,9],"n_estimators":[200,300,800],"random_state":[1842]},
    cv=5, n_iter=6
  )
```

```python
my_rsearch.fit(X_train2,y_train)
```

```python
my_results=pd.DataFrame(my_rsearch.cv_results_)
my_results[["param_max_features","param_n_estimators","mean_test_score","rank_test_score"]]
```

```python
#The param_max_features is the maximum number of features the randomized search chose to allow in each individual tree 
#in the Random Forest.
#The param_n_estimators represents the number of trees used in each iteration of the randomized search.
#The mean_test_score represents how well the resulting model does at predicting the y value (HardDrugs).
#Finally, the rank_test_score ranks out how well each randomized search did compared to each other. 
```
